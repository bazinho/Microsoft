```python
import glob
import csv
import os
import shutil
import array
import scipy.misc
import numpy as np
np.random.seed(1)
```

```python
def GetWidth(size):
    widthtable = [(10240, 32),
                  (30720, 64),
                  (61440, 128),
                  (102400, 256),
                  (204800, 384),
                  (512000, 512),
                  (1024000, 768)]
    maxwidth = 1024
    for w in widthtable:
        if size<w[0]:
            return w[1]
    return maxwidth
```

```python
asmdir = 'train'
```

```python
imagedir = 'AsmImages'
```

```python
if not os.path.exists(imagedir):
    os.makedirs(imagedir)
```

```python
flist = glob.glob(os.path.join(asmdir,"*.asm"))
```

```python
num_samples = len(flist)
print("Number of samples: %i" %(num_samples))
```

```python
flist
```

```python
count = 0
for fname in flist:
    with open(fname,"rb") as f:
        size = os.path.getsize(fname)
        width = GetWidth(size)
        a = array.array("B") # uint8 array
        a.fromfile(f,size)
        g = np.reshape(a[:size-(size%width)],(size//width,width))
        imgfile = os.path.join(imagedir,os.path.basename(fname).split('.')[0]+'.png')
        print("[%i] Converting file %s to %s" %(count,fname,imgfile))
        scipy.misc.imsave(imgfile,g)
        count += 1
        
print("Converted files: %i" %(count))
```

```python
trainLabelFile = "trainLabels.csv"
```

```python
with open(trainLabelFile) as f:
    labelDict = dict(filter(None, csv.reader(f)))
```

```python
labelDict
```

```python
y = np.zeros(num_samples)
pos = 0
for f in flist:
    fname = os.path.splitext(os.path.basename(f))[0]
    print("File: %s - Class: %s" %(fname,labelDict[fname]))
    y[pos] = int(labelDict[fname])
    pos += 1
```

```python
list_fams = ['Ramnit','Lollipop','Kelihos_ver3','Vundo','Simda','Tracur','Kelihos_ver1','Obfuscator.ACY','Gatak']
```

```python
imagedirbyfamily = 'AsmImagesByFamily'
```

```python
if not os.path.exists(imagedirbyfamily):
    os.makedirs(imagedirbyfamily)
    
for fam in list_fams:
    if not os.path.exists(os.path.join(imagedirbyfamily,fam)):
        os.makedirs(os.path.join(imagedirbyfamily,fam))
```

```python
count = 0
for f in flist:
    mwhash = os.path.splitext(os.path.basename(f))[0]
    fam = list_fams[int(labelDict[mwhash])-1]
    count += 1
    print("[%i] Hash: %s - Class: %s - Dir: %s" %(count,mwhash,labelDict[mwhash],fam))
    shutil.copy(os.path.join(imagedir,mwhash+".png"),os.path.join(imagedirbyfamily,fam,mwhash+".png"))
```
